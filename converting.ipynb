```python
#0 : dad
#1 : husband
translator = Translator()

name = []
Relation = []
House = []
age = []
gender = []
for i in f:
    
    name.append(translator.translate(f[i]["नाम"]).text)
    if "Father" in f[i]:
        m = '0-'+translator.translate(f[i]["Father"]).text
        Relation.append(m)
    elif "Husband" in f[i]:
        m = '1-'+translator.translate(f[i]["Husband"]).text
        Relation.append(m)
    House.append(translator.translate(f[i]["मकानसंख्या"]).text)
    age.append(translator.translate(f[i]["आयु"]).text)
    gender.append(translator.translate(f[i]["लिंग"]).text)
df = pd.DataFrame(list(zip(name,Relation,House,age,gender)),columns =['Names','Relation','House','age','gender'])

    
```
